# Extracción de datos para preguntas Streamlit

## Carga de datos

```python
import pandas as pd

data = pd.read_csv('nuevo_df.csv')

# Convertir todas las variables a tipo str
data = data.astype(str)

data = data.drop('Otro..especifique..1', axis=1)

data.head()

data['Características.del.implante']
```

## Filtramos 10 columnas

```python
columnas_seleccionadas = ['Tipo.de.Intervención.Quirúrgica', 'Alcohol', 'Medicación.Actual', 
                          'Otras.Drogas', 'Tipo.de.prótesis.sobre.implantes', 'Tipo.de.cirugía', 
                          'Implante.1...Defecto.tipo.I..infraóseo.', 'X.Qué.material.de.regeneración.ha.sido.utilizado.',
                          'Número.de.implantes', 'Características.del.implante']

df = data[columnas_seleccionadas]

df.head()
```

```python
import pandas as pd

# Suponiendo que df es tu DataFrame
for columna in df.columns:
    valores_unicos = df[columna].unique()
    print(f'Valores únicos en {columna}: {valores_unicos}\n')
```
