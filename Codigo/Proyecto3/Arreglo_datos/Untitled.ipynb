```python
import pandas as pd

# Supongamos que tienes un DataFrame llamado df y una lista de rangos de columnas seleccionadas
column_ranges = [(0, 928)]  # Ejemplo de rangos de columnas seleccionadas

# Función para transformar los valores numéricos a 1
def transform_values(x):
    if isinstance(x, (int, float)) and x != 0:  # Verifica si x es un número y no es 0
        return 1
    else:
        return x

# Aplicar la función a las columnas seleccionadas del DataFrame
for start, end in column_ranges:
    df.iloc[:, start:end+1] = df.iloc[:, start:end+1].applymap(transform_values)

    
df
```

```python
# Exportar el DataFrame a un archivo CSV
df.to_csv('combianted_Total.csv', index=False)
df

```
