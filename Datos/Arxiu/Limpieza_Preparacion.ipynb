## 1. Loading the data

At this stage, we load the different .csv files, and we will briefly display them.

The objective is to unify them into a single file that you can work with.

```python
import pandas as pd

df1 = pd.read_csv('Datos/CSV2/Hoja de Recogida de Datos 20172018.csv')
df2 = pd.read_csv('Datos/CSV2016/Hoja de Recogida de Datos.csv')
df3 = pd.read_csv('Datos/CSV3/Hoja de Recogida de Datos 20182019.csv')
df4 = pd.read_csv('Datos/CSV4/Hoja de Recogida de Datos 2016.csv')

df1.head()
```

well, like we see all the data are like same dimensions, we can to follow :)

## 2. Join

At this point, we going to Join all of the dataFrame only in one.

That moment are complety importat, we need to join the data as correct way to follow with the study.

For this, we will going to use the concat function, we only have 1 argument, **'axis=0 or axis=1'**, we use **1** for join the data as rows, **0** are for join data for columns.

```python
# Function concat (columns --> axis = 0)
df = pd.concat([df1,df2,df3,df4], axis=0)
df.head()
```

Wow, we did it, so now, we go for the next step.

## 3. Arreglar el nombre

At this point, we will go to use only the first row at data for variables names.

We will only change de name of colums that estar for **Unnamed:...**

```python
for col in df.columns:
    if col.startswith('Unnamed:'):
        df.rename(columns={col: df.iloc[0][col]}, inplace=True)
df = df.iloc[1:, :]
```

```python
df.head()
```

So, we change all the names, now the data are ready for **cleanig**.

## 4. Separation by fild of study

Finally we will to separate the big dataFrame in to diferent fild of study:

- Systemic pathology
- Pre-operation medication
- Post-operation medication
- Surgical tools

lets do it

```python
# Systemic pathology
Systemic_Pathology = df.iloc[:, 0:67]
Systemic_Pathology.to_csv('Systemic_Pathology.csv', index=False)

# Pre-operation medication (Tipo de intervencion)
Pre_operation_medication = df.iloc[:, 67:119]
Pre_operation_medication.to_csv('Pre_operation_medication.csv', index=False)

# Surgical tools (Tipo de intervencion)
Surgical_tools = df.iloc[:, 119:914]
Surgical_tools.to_csv('Surgical_tools.csv', index=False)

# Post-operation medication
Post_operation_medication = df.iloc[:, 914:950]
Post_operation_medication.to_csv('Post_operation_medication.csv', index=False)
```
